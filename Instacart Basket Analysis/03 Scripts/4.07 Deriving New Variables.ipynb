# 02. User-Defined Function "price_label"

```python
# define a price label function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# apply price label function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

# 04. For-Loops

```python
# find out which day most orders take place
ords_prods_merged['orders_day_of_the_week'].value_counts(dropna = False)
```

```python
# run the loop
result = []

for value in ords_prods_merged["orders_day_of_the_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# create a new column with dataframe = result
ords_prods_merged['busiest_day'] = result
```

```python
ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

## 02. create busiest and slowest days column

```python
# run the loop
result_2 = []

for value in ords_prods_merged["orders_day_of_the_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regular days")
```

```python
# create a new column with dataframe = result_2
ords_prods_merged['busiest_days'] = result_2
```

```python
ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

### Observations - Saturday and Sunday make up the count of busiest days while Friday, Thursday and Monday make up the regular days count - here we see that regular days and busy days are close to the same number. This means that lots of orders are being placed on Saturdays and Sundays.

## 04. create busiest period of day column

```python
# check frequency
ords_prods_merged['order_hour_of_day'].value_counts()
```

```python
# Create a for-loop if statement using the above breakdown
hour_of_day = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    hour_of_day.append("busiest time of day")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    hour_of_day.append("slowest time of day")
  else:
    hour_of_day.append("regularly busy time of day")
```

```python
# create a new column with dataframe = hour_of_day
ords_prods_merged['busiest_time_of_day'] = hour_of_day
```

```python
ords_prods_merged['busiest_time_of_day'].value_counts(dropna = False)
```

```python
ords_prods_merged.shape
```

### Observations - most orders (about 65%) happen at 9:00am - 4:00pm. About 4% of orders occur at 11:00pm - 6:00am. That leaves about 31% of orders occuring at 7:00 - 8:00am and 5:00 - 10:00pm.
